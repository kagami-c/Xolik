```python
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
from matplotlib_venn import venn3_unweighted
import sys


if len(sys.argv) <= 1:
    print 'Please specify input filename. First intra and then inter.'
    exit()


data = pd.read_csv(sys.argv[1], sep='\t', skiprows=[0],
                   names=['PSMId' ,'score', 'q-value', 'posterior_error_prob', 'peptide', 'proteinId#1', 'proteinId#2'])

def extract_scan_num(PSMId):
    words = PSMId.split('-')
    return int(words[1])

data['ScanNum'] = data['PSMId'].apply(extract_scan_num)

flags = []
for i in range(len(data['ScanNum'])):
    if i == 0:
        flags.append(0)
        continue
    current = data['ScanNum'][i]
    if data['ScanNum'][:i].isin([current]).any():
        flags.append(1)
    else:
        flags.append(0)

data['Flag'] = pd.Series(flags)

filtered = data[data['Flag'] != 1]
print('Total {} PSMs passing the threshold.'.format(len(filtered[filtered['q-value'] <= 0.05])))
filtered.to_csv(sys.argv[1] + '.dedup.csv', index=False)
```
